#### 7.	From superstore dataset, prepare the dataset to answer following question. Provide explanation on each step as well.


```python
# load file
superstore_order = pd.read_csv("/Users/kasinphatpam/Downloads/superstore_order.csv")
superstore_people = pd.read_csv("/Users/kasinphatpam/Downloads/superstore_people.csv")
superstore_return = pd.read_csv("/Users/kasinphatpam/Downloads/superstore_return.csv")
print(superstore_order)
```

7.1 Observe the weekly sales of each product 

```python
# convert string to pandas time stamp
superstore_order['Order Date'] = pd.to_datetime(superstore_order['Order Date'], format='%d/%m/%Y')
superstore_order['Ship Date'] = pd.to_datetime(superstore_order['Ship Date'], format='%d/%m/%Y')
# sort by order date
lating = superstore_order.sort_values(by='Order Date')
# find maximum of date (get lastest date of order)and filter out orders that have passed more than 7 days.
superstore_order['Product ID'][(lating['Order Date'].max()-superstore_order['Order Date'])/ np.timedelta64(1, 'D') <= 7].value_counts()
```

7.2 Find out which customers have returned products at least once 

```python
# sorted data from customer name
superstore_order = superstore_order.sort_values('Customer Name')
# check order id that have a list in another file or not and add 'returned' column
superstore_order = superstore_order.assign(returned=superstore_order['Order ID'].isin(superstore_return['Order ID']).astype(int))
print(superstore_order.filter(items=['Order ID','returned']))
```

```python
# frequency of returned product
superstore_order['Customer Name'][superstore_order.returned == 1].value_counts()
```

7.3 Sales to profit ratio of each subcategory

```python
# convert type of series as float
superstore_order['Sales'] = superstore_order['Sales'].astype(float)
# group product as sub-category and calculate sum of profit
profit = superstore_order.groupby(['Sub-Category'])['Profit'].sum(numeric_only=True)
# find frequency of each subcategory
frequency = superstore_order.groupby(['Sub-Category'])['Sub-Category'].count()
# convert 2 series to 1 dataframe
ratio = pd.concat([frequency, profit], axis=1)
# change column name
ratio = ratio.rename(columns={'Sub-Category': 'Frequency'})
# calculating ratio
ratio['Ratio'] = ratio['Frequency']/ratio['Profit']
print(ratio)
```

7.4 Orders that takes more than 5 days to prepare for shipping

```python
# Find the distance between the order date and the delivery date and add a column. 'diff date' to store results
superstore_order['diff day'] = (superstore_order['Ship Date'] - superstore_order['Order Date']) / np.timedelta64(1, 'D')
# print data with filter and show only order id, order date, ship date and diff day column
print(superstore_order[superstore_order['diff day'] > 5].filter(items=['Order ID','Order Date','Ship Date','diff day']).sort_values('diff day'))
```

```python
# the number of orders in each state
superstore_order.groupby(['Ship Mode','State']).size()
```

### Matplotlib 

From superstore dataset, plot visualization to answer following question using appropriate visual element. 

```python
import matplotlib.pyplot as plt
```

#### 1.	Number of customer who has returned product at least once and those who never returned products. 

```python
customer_retuned = superstore_order['Customer Name'][superstore_order.returned == 1].drop_duplicates().count()
customer_never_returned = superstore_order['Customer Name'][superstore_order.returned == 0].drop_duplicates().count()
plt.bar(['Returned', 'Never Returned'], [customer_retuned, customer_never_returned], color=['SlateBlue', 'Violet'])
plt.show()
```

#### 2.	Weekly sales of each product category. 

```python
weekly_category = superstore_order['Category'][(lating['Order Date'].max()-superstore_order['Order Date'])/ np.timedelta64(1, 'D') <= 7].value_counts()
# lastest week
print('Lastest week')
weekly_category.plot(kind='barh')
plt.show()

# overview sold amount
print('Overview sold amount (Quantity)')
weekly_sales = superstore_order.groupby(['Category', pd.Grouper(key='Order Date', freq='W')])['Quantity'].sum().reset_index()

plt.figure(figsize=(20, 6))
for category, group in weekly_sales.groupby('Category'):
    plt.plot(group['Order Date'], group['Quantity'], label=category)
plt.legend()
plt.xlabel('Date')
plt.ylabel('Sold amount')
plt.show()

# overview income
print('Overview income')
superstore_order['Total Sales'] = ((superstore_order['Sales'] * superstore_order['Quantity']) - superstore_order['Discount'])
weekly_sales = superstore_order.groupby(['Category', pd.Grouper(key='Order Date', freq='W')])['Total Sales'].sum().reset_index()

plt.figure(figsize=(20, 6))
for category, group in weekly_sales.groupby('Category'):
    plt.plot(group['Order Date'], group['Total Sales'], label=category)
plt.legend()
plt.xlabel('Date')
plt.ylabel('Sold amount')
plt.show()
```

#### 3.	Distribution of how many times a customer purchased product from this store. 

```python
plt.hist(superstore_order['Customer Name'].value_counts(), align='left', rwidth=0.8,color='SlateBlue')
print(superstore_order['Customer Name'].value_counts())
plt.ylabel('Number of customer')
plt.xlabel('Frequency purchased')
plt.show()
```

#### 4.	Comparative statistical distribution of profit per sale unit between each sub-category in furniture category. 


```python
# Retrieve only the furniture category and divide data by subcategories
sub_category = superstore_order[superstore_order['Category'] == 'Furniture'].groupby('Sub-Category')
# find the number of product
sub_category.size().plot(kind='barh')
plt.show()

# calculating ratio (sum of profit / number of product that sold)
df_profit_per_sale_unit = sub_category['Profit'].sum() / sub_category.size()
# ploting graph
df_profit_per_sale_unit.plot(kind='bar')
# setting label of graph
plt.xlabel('Sub Category')
plt.ylabel('Profit per Sale')
plt.show()
```
